# In IPL many players have registered their mark by giving amazing performances consisitently.

From the given datasets-

-Get Suresh Raina's individual scores from each match from the seasons 2008,2010,2011,2012,2013,2015 and
2019 respectively.

-Now get Suresh Raina's individual scores from each match in home city(Chennai) and individual scores from away cities((i.e) other than Chennai).

-Now,create a random sample of 20 data with seed value of 2.

-From the sample data sets perform t-test(paired) on individual runs scored by raina on home city and in
away cities

-Print the p-value

```python
seasons = [2008,2010,2011,2012,2013,2015]
seasons_match_ids_home = matche[(matche.season.isin(seasons)) & (matche.city == 'Chennai')].id.values.tolist()
seasons_match_ids_away = matche[(matche.season.isin(seasons)) & (matche.city != 'Chennai')].id.values.tolist()
```

```python
relevant_deli_home = deli[deli.match_id.isin(seasons_match_ids_home) & (deli.batsman ==  'SK Raina')]
relevant_deli_away = deli[deli.match_id.isin(seasons_match_ids_away) & (deli.batsman ==  'SK Raina')]
```

```python
relevant_deli_home_runs = relevant_deli_home.groupby('match_id').batsman_runs.sum().values.tolist()
relevant_deli_away_runs = relevant_deli_away.groupby('match_id').batsman_runs.sum().values.tolist()
```

```python
import random
random.seed(2)
relevant_deli_home_runs_sample = random.sample(relevant_deli_home_runs, 20)
relevant_deli_away_runs_sample = random.sample(relevant_deli_away_runs, 20)
```

```python
import scipy.stats as stats
t_stat, p_val = stats.ttest_ind(relevant_deli_away_runs_sample, relevant_deli_home_runs_sample)
print('t_test_statistics is',round(t_stat,2))
print('p_value is', round(p_val,2))
```

# In IPL strike rate is important where batsman having higher strike rate helps team to get more runs.

From the given datasets-


-Get Rohit Sharma's individual strike rates from each match where he played for Mumbai Indians.

-Now from the above data get Rohit Sharma's individual strike rates in home city(Mumbai) and individual  
strike rates from away cities((i.e) other than Mumbai).

-Now, create a random sample of 20 data with seed value of 2.

-From the sample data sets perform t-test(paired) on individual runs scored by rohit on home city and in     
away cities

-Formula for calculating strike rate - (batsmen runs/ball faced)*100

-Print p value

```python
relevant_match_ids = list(set(deli[(deli.batsman=='RG Sharma') & (deli.batting_team=='Mumbai Indians')].match_id.values.tolist()))

relevant_matches_home = matche[(matche.id.isin(relevant_match_ids)) & (matche.city=="Mumbai")].id.values.tolist()
relevant_matches_away = matche[(matche.id.isin(relevant_match_ids)) & (matche.city!="Mumbai")].id.values.tolist()
```

```python
relevant_matches_home_deli = deli[(deli.match_id.isin(relevant_matches_home)) & (deli.batsman == 'RG Sharma')]
relevant_matches_away_deli = deli[(deli.match_id.isin(relevant_matches_away)) & (deli.batsman == 'RG Sharma')]
```

```python
home_match_balls_faced = relevant_matches_home_deli.groupby('match_id').batsman.count().values.tolist()
home_match_runs_scored = relevant_matches_home_deli.groupby('match_id').batsman_runs.sum().values.tolist()

away_match_balls_faced = relevant_matches_away_deli.groupby('match_id').batsman.count().values.tolist()
away_match_runs_scored = relevant_matches_away_deli.groupby('match_id').batsman_runs.sum().values.tolist()
```

```python
home_sr = [ round(((home_match_runs_scored[i]/home_match_balls_faced[i])*100),2) for i in range(len(home_match_balls_faced))]
away_sr = [ round(((away_match_runs_scored[i]/away_match_balls_faced[i])*100),2) for i in range(len(away_match_balls_faced))]
```

```python
import random
random.seed(2)
home_sr_sample = random.sample(home_sr, 20)
away_sr_sample = random.sample(away_sr, 20)

t_stat, p_val = stats.ttest_ind(home_sr_sample, away_sr_sample)
print('t_test_statistics is',round(t_stat,2))
print('p_value is', round(p_val,2))
```
